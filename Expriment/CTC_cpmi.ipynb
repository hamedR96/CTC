```python
import math
import os
import pickle
from tqdm import tqdm

ntm_results_path = "./NTM/results/"

coherence_result = "ctc_result_tweet_rest.txt"

with open("./cpmi/out_tweets", "rb") as fp:   # Unpickling
   outs = pickle.load(fp)

with open("./cpmi/sentences_tweets", "rb") as fp:   # Unpickling
   sentences = pickle.load(fp)

with open(coherence_result, "w") as f:
    for filename in os.listdir(ntm_results_path):

        name=filename.split("_")
        if 'topics' in name and "tweets" in name:
            with open("NTM/results/"+str(filename), "rb") as fp:
                topics = pickle.load(fp)
            sentences_score=0
            for i, sentence in enumerate(tqdm(sentences)):
                topics_score=0
                for topic in topics:
                    topic_score=0
                    for word1 in topic:
                        for word2 in topic:
                            if word1 in sentence and word2 in sentence:
                                w1=sentence.index(word1)
                                w2=sentence.index(word2)
                                topic_score+=outs[0][str(i)][w1][w2]/2
                    topics_score+=topic_score/45
                sentences_score+=topics_score/len(topics)
            f.write(f"{filename}, {sentences_score*math.exp(6)/len(sentences)}\n")
    f.write("The experiment is successfully done!")
```

```python
import math
import os
import pickle
from tqdm import tqdm

ntm_results_path = "./NTM/results/"

coherence_result = "ctc_result_20ng.txt"


with open(coherence_result, "w") as f:
    for filename in os.listdir(ntm_results_path):
        name=filename.split("_")
        if 'topics' in name and "20newsgroup" in name:
            with open("NTM/results/"+str(filename), "rb") as fp:
                topics = pickle.load(fp)
            sentences_score=0
            len_sentences=0
            for p in range(1,8):
                with open("./cpmi/sentences_20_part"+str(p), "rb") as fp:   # Unpickling
                    sentences = pickle.load(fp)

                with open("./cpmi/outs_20_part"+str(p), "rb") as fp:   # Unpickling
                    outs = pickle.load(fp)
                len_sentences+=len(sentences)
                for i, sentence in enumerate(tqdm(sentences)):
                    topics_score=0
                    for topic in topics:
                        topic_score=0
                        for word1 in topic:
                            for word2 in topic:
                                if word1 in sentence and word2 in sentence:
                                    w1=sentence.index(word1)
                                    w2=sentence.index(word2)
                                    topic_score+=outs[0][str(i)][w1][w2]/2
                        topics_score+=topic_score/45
                    sentences_score+=topics_score/len(topics)

            f.write(f"{filename}, {sentences_score*math.exp(6)/len_sentences}\n")
    f.write("The experiment is successfully done!")
```
